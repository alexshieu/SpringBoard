```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Initial Analysis
filename = 'Consumer_Complaints.csv'
df = pd.read_csv(filename, parse_dates=True, index_col='Date received')
df.columns
df.shape
df.info()
df.describe()
```

```python
#Column Value Count
column_value_count = []
for column_label in df.columns:
    column_value_count.append(df[column_label].value_counts(dropna=False))
print(column_value_count)
```
