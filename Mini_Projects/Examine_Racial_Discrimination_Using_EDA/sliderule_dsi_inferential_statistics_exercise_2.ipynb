# Examining Racial Discrimination in the US Job Market

### Background
Racial discrimination continues to be pervasive in cultures throughout the world. Researchers examined the level of racial discrimination in the United States labor market by randomly assigning identical résumés to black-sounding or white-sounding names and observing the impact on requests for interviews from employers.

### Data
In the dataset provided, each row represents a resume. The 'race' column has two values, 'b' and 'w', indicating black-sounding and white-sounding. The column 'call' has two values, 1 and 0, indicating whether the resume received a call from employers or not.

Note that the 'b' and 'w' values in race are assigned randomly to the resumes when presented to the employer.

1. What test is appropriate for this problem? Does CLT apply?
Binomial distribution appears to be best as we have number of call backs which relates to trials with a probability of success given by call backs due to black or white sounding name. The call backs are independent of one another as the resumes are not dependent on one another and  the same number of black sounding and white sounding named resumes were sent. CLT therefore applies.

2. What are the null and alternate hypotheses?
The null hypothesis is that having a black or white sounding name does not impact call back rates from employers. The alternate hypothesis is that having a black sounding name does impact call back rates from emplyoers compared to having a white sounding name

```python
# Under null hypothesis with black and white sounding names being the same then
# the chance is 50/50 for both groupings with 392 trials. The number of resumes
# submitted were the same for both groupings so that is eliminated as a variable
samples = np.random.binomial(392, 0.5, size=10000)

conf_int = np.percentile(samples, [2.5, 97.5])
print('95% confidence interval is: ',conf_int)
```
